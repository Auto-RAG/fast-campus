# PDF Library 파싱 실습

## Setting

```python
from glob import glob
import os

import pandas as pd

# Path Setting
current_dir = os.getcwd()
root_dir = os.path.join(current_dir, '..', '..')
data_dir = os.path.join(root_dir, 'data', 'parse_single')
glob_path = os.path.join(data_dir, '*')

pdf_path = glob(glob_path)[0]
pdf_path
```

## 1. PyPDFium2

```python
from langchain_community.document_loaders import PyPDFium2Loader

loader = PyPDFium2Loader(pdf_path)
load = loader.load()
pypdfium2_result = [page.page_content for page in load]
```

```python
print(pypdfium2_result[3])
```

## 2. PyMuPDF

```python
import fitz
    
doc = fitz.open(pdf_path)
pymupdf_result = [page.get_text() for page in doc]
```

```python
print(pymupdf_result[3])
```

## 3. PyPDF

```python
from pypdf import PdfReader

reader = PdfReader(pdf_path)
pages = reader.pages

pypdf_result = [page.extract_text() for page in pages]
```

```python
print(pypdf_result[3])
```

## 4. PDFMiner

```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LTTextContainer

def extract_text_by_page(pdf_path):
    for page_layout in extract_pages(pdf_path):
        page_text = ""
        for element in page_layout:
            if isinstance(element, LTTextContainer):
                page_text += element.get_text()
        yield page_text
        
pdfminer_result = [page for page in extract_text_by_page(pdf_path)]
```

```python
print(pdfminer_result[3])
```

## 5. PDFPlumber

```python
import pdfplumber
    
pdf = pdfplumber.open(pdf_path)
pages = pdf.pages
pdfplumber_result = [page.extract_text() for page in pages]
```

```python
print(pdfplumber_result[3])
```

# Use AutoRAG

```python
from autorag.parser import Parser

multiple_glob_path = os.path.join(root_dir, 'data', 'parse_multi', '*')
project_dir = os.path.join(root_dir, 'autorag_project', 'parse', 'pdf_library')
parser = Parser(data_path_glob=multiple_glob_path, project_dir=project_dir)
```

```python
yaml_path = os.path.join(root_dir, 'config', 'parse', 'pdf_library.yaml')

parser.start_parsing(yaml_path)
```
